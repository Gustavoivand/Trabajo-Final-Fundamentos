```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error , mean_squared_error
```

```python
import pandas as pd
df = pd.read_csv("dataset.csv")

X = df.drop(columns = ["Drift" , "MasaPropia","Px"])

y1 = df['Drift']
y2 = df['MasaPropia']
```

```python
X_train, X_test , y1_train , y1_test = train_test_split(X, y1, train_size = 0.8, shuffle = True, random_state = 42 )
X_train, X_test , y2_train , y2_test = train_test_split(X, y2, train_size = 0.8, shuffle = True, random_state = 42 )

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
numerical_pipeline = Pipeline([
    ('scaler', StandardScaler())])

categorical_pipeline = Pipeline([
    ('one_hot_encoder', OneHotEncoder())])

numerical_features = df.drop(columns = ["Drift" , "MasaPropia","Px"]).select_dtypes(exclude = ["object"]).columns
categorical_features = df.drop(columns = ["Drift" , "MasaPropia","Px"]).select_dtypes(include = ["object"]).columns
```

```python
preprocessor = ColumnTransformer([
    ('num', numerical_pipeline, numerical_features),
    ('cat', categorical_pipeline, categorical_features)], remainder='passthrough')
```

```python
model1 = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor',
     RandomForestRegressor(n_estimators=100))])
model2 = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor',
     RandomForestRegressor(n_estimators=100))])
```

```python
model1.fit(X_train, y1_train)
model2.fit(X_train, y2_train)
```

```python
mean_absolute_percentage_error(model1.predict(X_test), y1_test)#Drift
```

```python
mean_absolute_percentage_error(model2.predict(X_test), y2_test)#MasaPropia
```

```python
import pickle

# File path where the model will be saved
model_file_path1 = 'Drift.pkl'
model_file_path2 = 'MasaPropia.pkl'

# Save the model to the pkl file
with open(model_file_path1, 'wb') as f:
    pickle.dump(model1, f)

# Save the model to the pkl file
with open(model_file_path2, 'wb') as f:
    pickle.dump(model2, f)
```
